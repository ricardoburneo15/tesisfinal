# Aplicar Eliminaci칩n de Outliers por Desviaci칩n Est치ndar

```python
# Funci칩n para eliminar outliers por desviaci칩n est치ndar (췀3픢)
def eliminar_outliers_std(df, columnas_numericas, num_std=3):
    """
    Elimina registros donde los valores est치n fuera de 췀3 desviaciones est치ndar

    Par치metros:
    - df: DataFrame original
    - columnas_numericas: lista de columnas num칠ricas a evaluar
    - num_std: n칰mero de desviaciones est치ndar (default=3)

    Retorna:
    - df_limpio: DataFrame sin outliers
    - info_eliminados: diccionario con informaci칩n de registros eliminados
    """
    df_limpio = df.copy()
    registros_iniciales = len(df_limpio)
    info_eliminados = {}

    for columna in columnas_numericas:
        # Calcular media y desviaci칩n est치ndar
        media = df_limpio[columna].mean()
        std = df_limpio[columna].std()

        # Calcular l칤mites
        limite_inferior = media - (num_std * std)
        limite_superior = media + (num_std * std)

        # Contar registros que se eliminar치n
        outliers = df_limpio[(df_limpio[columna] < limite_inferior) |
                             (df_limpio[columna] > limite_superior)]

        info_eliminados[columna] = {
            'media': media,
            'std': std,
            'limite_inferior': limite_inferior,
            'limite_superior': limite_superior,
            'registros_eliminados': len(outliers)
        }

        # Filtrar datos dentro del rango
        df_limpio = df_limpio[(df_limpio[columna] >= limite_inferior) &
                              (df_limpio[columna] <= limite_superior)]

    registros_finales = len(df_limpio)
    total_eliminados = registros_iniciales - registros_finales

    print(f"\n=== ELIMINACI칍N POR DESVIACI칍N EST츼NDAR (췀{num_std}픢) ===")
    print(f"Registros iniciales: {registros_iniciales:,}")
    print(f"Registros finales: {registros_finales:,}")
    print(f"Total eliminados: {total_eliminados:,} ({total_eliminados/registros_iniciales*100:.2f}%)")
    print("\nDetalle por variable:")
    for col, info in info_eliminados.items():
        print(f"  {col}:")
        print(f"    Media: {info['media']:,.2f}")
        print(f"    Std: {info['std']:,.2f}")
        print(f"    L칤mite inferior: {info['limite_inferior']:,.2f}")
        print(f"    L칤mite superior: {info['limite_superior']:,.2f}")

    return df_limpio, info_eliminados

# Definir las variables num칠ricas sobre las que aplicar la eliminaci칩n de outliers
variables_financieras = [
    'activos',
    'patrimonio',
    'utilidad_neta',
    'ingresos_ventas',
    'ingresos_totales',
    'utilidad_an_imp',
    'liquidez_corriente',
    'gastos_financieros',
    'end_activo',
    'end_corto_plazo',
    'end_largo_plazo'
]

#generar el df_limpio sin outliers
df_limpio, info_std = eliminar_outliers_std(df, variables_financieras, num_std=3)
```

# Aplicar Winsorizaci칩n para Manejo de Outliers Extremos

```python
from scipy.stats.mstats import winsorize

def aplicar_winsorization(df, columnas, limites=[0.01, 0.01]):
    """
    Reemplaza los outliers con los valores de los percentiles espec칤ficos.

    Par치metros:
    - df: DataFrame de trabajo.
    - columnas: Lista de variables financieras a winsorizar.
    - limites: [l칤mite_inferior, l칤mite_superior] ej. 0.01 es el percentil 1 y 99.
    """
    df_winsor = df.copy()

    for col in columnas:
        # Aplicamos winsorize de scipy
        # limits=[0.01, 0.01] indica que el 1% m치s bajo y el 1% m치s alto se ajustan
        df_winsor[col] = winsorize(df_winsor[col], limits=limites)

    print(f"九 Winsorizaci칩n completada en {len(columnas)} variables.")
    return df_winsor

# Definimos las variables que detectamos con problemas (especialmente la de los billones)
variables_a_corregir = [
    'utilidad_an_imp', 'activos', 'patrimonio', 'utilidad_neta',
    'ingresos_totales', 'liquidez_corriente', 'end_activo'
]

# Aplicamos la t칠cnica al percentil 1% y 99%
df_limpio_win = aplicar_winsorization(df_limpio, variables_a_corregir, limites=[0.01, 0.01])

# Verificamos los nuevos estad칤sticos
print(df_limpio_win[variables_a_corregir].describe().round(2))
```

# Limpieza del DataFrame Basada en L칩gica Financiera

```python
# ============================================================================
# LIMPIEZA POR L칍GICA FINANCIERA
# ============================================================================

def limpiar_por_logica_financiera(df_limpio):
    """
    Aplica filtros basados en l칩gica financiera y de negocio
    para eliminar casos at칤picos o errores de datos
    """
    df_clean = df_limpio.copy()
    registros_iniciales = len(df_clean)

    print("="*80)
    print("LIMPIEZA POR L칍GICA FINANCIERA")
    print("="*80)
    print(f"Registros iniciales: {registros_iniciales:,}\n")

    # 1. Eliminar empresas sin activos o con activos muy peque침os
    print("1勇  Filtro: Activos > 0")
    antes = len(df_clean)
    df_clean = df_clean[df_clean['activos'] > 0]
    eliminados = antes - len(df_clean)
    print(f"   Eliminados: {eliminados:,} empresas sin activos")

    # 2. Eliminar empresas sin operaci칩n (ingresos = 0)
    #print("\n2勇  Filtro: Ingresos Totales > 0")
    #antes = len(df_clean)
    #df_clean = df_clean[df_clean['ingresos_totales'] > 0]
    #eliminados = antes - len(df_clean)
    #print(f"   Eliminados: {eliminados:,} empresas sin ingresos")

    # 3. Eliminar indicadores de endeudamiento imposibles
    print("\n3勇  Filtro: Endeudamiento dentro de rangos l칩gicos")

    # end_activo debe estar entre 0 y 2 (0% a 100%)
    antes = len(df_clean)
    df_clean = df_clean[(df_clean['end_activo'] >= 0) & (df_clean['end_activo'] <= 2.0)]
    eliminados = antes - len(df_clean)
    print(f"   end_activo fuera de rango [0,1]: {eliminados:,} eliminados")

    # end_corto_plazo debe estar entre 0 y 1
    antes = len(df_clean)
    df_clean = df_clean[(df_clean['end_corto_plazo'] >= 0) & (df_clean['end_corto_plazo'] <= 1)]
    eliminados = antes - len(df_clean)
    print(f"   end_corto_plazo fuera de rango [0,1]: {eliminados:,} eliminados")

    # end_largo_plazo debe estar entre 0 y 1
    antes = len(df_clean)
    df_clean = df_clean[(df_clean['end_largo_plazo'] >= 0) & (df_clean['end_largo_plazo'] <= 1)]
    eliminados = antes - len(df_clean)
    print(f"   end_largo_plazo fuera de rango [0,1]: {eliminados:,} eliminados")

    # 4. Liquidez corriente razonable (0 a 20)
    print("\n4勇  Filtro: Liquidez Corriente razonable [0, 20]")
    antes = len(df_clean)
    df_clean = df_clean[(df_clean['liquidez_corriente'] >= 0) &
                        (df_clean['liquidez_corriente'] <= 20)]
    eliminados = antes - len(df_clean)
    print(f"   Eliminados: {eliminados:,} empresas con liquidez extrema")

    # 5. Eliminar empresas sin ingresos por ventas (ingresos = 0)
    print("\n2勇  Filtro: ingresos_ventas  > 0")
    antes = len(df_clean)
    df_clean = df_clean[df_clean['ingresos_ventas'] > 0]
    eliminados = antes - len(df_clean)
    print(f"   Eliminados: {eliminados:,} empresas sin ingresos")

    #8. Eliminar empresas sin pasivo total (pasivos = 0)
    print("\n2勇  Filtro: pasivo total  > 0")
    antes = len(df_clean)
    df_clean = df_clean[df_clean['pasivo_total'] > 0]
    eliminados = antes - len(df_clean)
    print(f"   Eliminados: {eliminados:,} empresas sin pasivo total")

    #9. Eliminar empresas sin deuda_total (deuda total = 0)
    #print("\n2勇  Filtro: deuda total  > 0")
    #antes = len(df_clean)
    #df_clean = df_clean[df_clean['deuda_total'] > 0]
    #eliminados = antes - len(df_clean)
    #print(f"   Eliminados: {eliminados:,} empresas sin deuda total")

    # Resumen final
    registros_finales = len(df_clean)
    total_eliminados = registros_iniciales - registros_finales

    print("\n" + "="*80)
    print("RESUMEN DE LIMPIEZA")
    print("="*80)
    print(f"Registros iniciales:  {registros_iniciales:,}")
    print(f"Registros finales:    {registros_finales:,}")
    print(f"Total eliminados:     {total_eliminados:,} ({total_eliminados/registros_iniciales*100:.2f}%)")
    print(f"Datos conservados:    {registros_finales/registros_iniciales*100:.2f}%")
    print("="*80)

    return df_clean

def estadisticas_comparativas(df_original, df_limpio):
    """
    Compara estad칤sticas antes y despu칠s de la limpieza
    """
    print("\n" + "="*80)
    print("COMPARACI칍N: ANTES vs DESPU칄S DE LIMPIEZA")
    print("="*80)

    variables = ['activos', 'patrimonio', 'utilidad_neta', 'ingresos_totales',
                 'liquidez_corriente', "ingresos_ventas", 'pasivo_total' ]

    comparacion = pd.DataFrame({
        'Variable': variables,
        'Media_Antes': [df_original[v].mean() for v in variables],
        'Media_Despu칠s': [df_limpio[v].mean() for v in variables],
        'Mediana_Antes': [df_original[v].median() for v in variables],
        'Mediana_Despu칠s': [df_limpio[v].median() for v in variables],
        'Std_Antes': [df_original[v].std() for v in variables],
        'Std_Despu칠s': [df_limpio[v].std() for v in variables]
    })

    print("\n", comparacion.to_string(index=False))

    return comparacion


# ============================================================================
# PIPELINE COMPLETO DE LIMPIEZA
# ============================================================================

def pipeline_limpieza_completo(df_limpio):
    """
    Ejecuta todo el pipeline de limpieza en secuencia
    """
    print("\n" + "游댢 "*20)
    print("PIPELINE COMPLETO DE LIMPIEZA DE DATOS")
    print("游댢 "*20 + "\n")

    # Guardar estad칤sticas originales
    print("游늵 Estad칤sticas originales guardadas...\n")

    # Paso 1: Limpieza por l칩gica financiera
    df_step1 = limpiar_por_logica_financiera(df_limpio)

    # Paso 2: An치lisis de outliers (ya no se llama a analizar_outliers_z2)

    # Paso 3: Decisi칩n sobre outliers (No se incluye la secci칩n de Z2 Score)

    # Por defecto, NO eliminar outliers IQR
    df_final = df_step1.copy()

    # Paso 4: Comparaci칩n final
    print("\n")
    estadisticas_comparativas(df_limpio, df_final)

    # Paso 5: Reporte final
    print("\n" + "="*80)
    print("九 LIMPIEZA COMPLETADA")
    print("="*80)
    print(f"Dataset final: {len(df_final):,} registros")
    print(f"Reducci칩n total: {(1 - len(df_final)/len(df_limpio))*100:.2f}%")

    return df_final


# ============================================================================
# EJEMPLO DE USO
# ============================================================================

# Ejecutar pipeline completo
df_final_limpio = pipeline_limpieza_completo(df_limpio)

# Si decides eliminar outliers IQR adicionales:
# df_final_limpio = eliminar_outliers_iqr(df_final_limpio, aplicar_a_z2=True)

#Guardar resultado
#df_final_limpio.to_csv('datos_finales_limpios.csv', index=False)

print("\n游 Para guardar: df_final_limpio.to_csv('datos_finales_limpios.csv', index=False)")
```

# Calcular los ratios para el modelo Z2 Score

```python
#generar el ratio x1
df_final_limpio['X1_liquidez'] = df_final_limpio['capital_trabajo'] / df_final_limpio['activos']

#generar el ratio x2
#Usar utilidad neta como proxy
df_final_limpio['X2_reinversi칩n'] = df_final_limpio['utilidad_neta'] / df_final_limpio['activos']

#generar el ratio x3
# EBIT = Utilidad antes de impuestos + Gastos financieros
df_final_limpio['ebit'] = df_final_limpio['utilidad_an_imp'] + df_final_limpio['gastos_financieros']
df_final_limpio['X3_rendimiento'] = df_final_limpio['ebit'] / df_final_limpio['activos']

#generar el ratio x2
# Usar pasivos totales ya calculado
df_final_limpio['X4_solvencia'] = df_final_limpio['patrimonio'] / df_final_limpio['pasivo_total']
```

# Calcular el Z2 Score (Altman)

```python
df_final_limpio['Z2 Score'] = 6.56 * df_final_limpio['X1_liquidez'] + 3.62 * df_final_limpio['X2_reinversi칩n'] + 6.72 * df_final_limpio['X3_rendimiento'] + 1.05 * df_final_limpio['X4_solvencia']
```

# Calcular los ratios para el modelo CA SCORE

```python
# X1 = Solidez Financiera
print("\n游늵 X1_CA_Score Solidez Financiera = Patrimonio (Capital Contable) / Activo total")
df_final_limpio['X1_CA_Score'] = df_final_limpio['patrimonio'] / df_final_limpio['activos']

# X2 = rentabilidad financiera
print("\n游늵 X2_CA_Score rentabilidad financiera  = ebit / Activo Total")
df_final_limpio['X2_CA_Score'] = df_final_limpio['ebit'] / df_final_limpio['activos']

# X3 = efectividad en la administraci칩n
print("\n游늵 X3_CA_Score(Estructura de Deuda) = ingresos ventas / activos")
df_final_limpio['X3_CA_Score'] = df_final_limpio['ingresos_ventas'] / df_final_limpio['activos']
```

# Calcular los ratios para el modelo Springate

```python
# ============================================================================
# C츼LCULO DE RATIOS - MODELO SPRINGATE
# ============================================================================

print("="*80)
print("CALCULANDO RATIOS PARA MODELO SPRINGATE")
print("="*80)

# A = Capital de Trabajo / Activo Total
print("\n游늵 A_springate (Capital de Trabajo) = Capital de Trabajo / Activo Total")
df_final_limpio['A_springate'] = df_final_limpio['capital_trabajo'] / df_final_limpio['activos']

# B = EBIT / Activo Total
print("\n游늵 B_springate (Rentabilidad) = EBIT / Activo Total")
df_final_limpio['B_springate'] = df_final_limpio['ebit'] / df_final_limpio['activos']

# C = Utilidad antes de Impuestos / Pasivo Corriente
print("\n游늵 C_springate (Solvencia) = Utilidad antes de Impuestos / Pasivo Corriente")
df_final_limpio['C_springate'] = df_final_limpio['utilidad_an_imp'] / df_final_limpio['pasivo_corriente']

# D = Ventas / Activo Total
print("\n游늵 D_springate (Rotaci칩n) = Ventas / Activo Total")
df_final_limpio['D_springate'] = df_final_limpio['ingresos_ventas'] / df_final_limpio['activos']

print("\n九 Ratios de Springate calculados exitosamente")

# ============================================================================
# C츼LCULO DEL Z-SCORE SPRINGATE
# ============================================================================

print("\n" + "="*80)
print("CALCULANDO Z-SCORE SPRINGATE")
print("="*80)

# F칩rmula: Z = 1.03(A) + 3.07(B) + 0.66(C) + 0.40(D)
df_final_limpio['Springate_Score'] = (1.03 * df_final_limpio['A_springate'] +
                                      3.07 * df_final_limpio['B_springate'] +
                                      0.66 * df_final_limpio['C_springate'] +
                                      0.40 * df_final_limpio['D_springate'])

print("\n九 Springate Score calculado")
print("\nF칩rmula: Z = 1.03(A) + 3.07(B) + 0.66(C) + 0.40(D)")

# Estad칤sticas
print("\n游늵 Estad칤sticas Springate Score:")
print(df_final_limpio['Springate_Score'].describe())

# Clasificaci칩n seg칰n Springate
# Criterio: Z > 0.862 = Empresa saludable
#           Z \u2264 0.862 = Riesgo de quiebra

print("\n" + "="*80)
print("CLASIFICACI칍N SPRINGATE")
print("="*80)
print("\nCriterio de clasificaci칩n:")
print("  Z > 0.862:  Empresa saludable (No quiebra)")
print("  Z \u2264 0.862:  Riesgo de quiebra (Quiebra)")

# Variable dummy
df_final_limpio['springate_dummy'] = (df_final_limpio['Springate_Score'] > 0.862).astype(int)

# Variable categ칩rica
df_final_limpio['categoria_springate'] = df_final_limpio['springate_dummy'].map({
    0: 'quiebra',
    1: 'no quiebra'
})

# Distribuci칩n
print("\n游늵 Distribuci칩n springate_dummy:")
print(df_final_limpio['springate_dummy'].value_counts().sort_index())

print("\n游늵 Distribuci칩n categoria_springate:")
print(df_final_limpio['categoria_springate'].value_counts())

print("\n九 Modelo Springate completado exitosamente!")
```

# CREACI칍N DE VARIABLE DUMMY (ALTMAN Z2 SCORE Y CA SCORE)

```python
print("\n" + "="*80)
print("游꿢 PASO 3: C치lculo del CA Score")
print("="*80)

# Coeficientes del modelo
b1 = 4.5913
b2 = 4.5080
b3 = 0.3936
b4 = 2.7616

print(f"\nF칩rmula: CA = {b1}칑X1 + {b2}칑X2 + {b3}칑X3 - {b4}")
print(f"\nCoeficientes:")
print(f"  4:                 {b4:>10.5f}")
print(f"  1 (X1 - Liquidez): {b1:>10.5f}")
print(f"  2 (X2 - Reinv.):   {b2:>10.5f}")
print(f"  3 (X3 - Est.Deuda):{b3:>10.5f}")

# Calcular CA_score
df_final_limpio['CA_Score'] = (b1 * df_final_limpio['X1_CA_Score'] +
                     b2 * df_final_limpio['X2_CA_Score'] +
                     b3 * df_final_limpio['X3_CA_Score']) - b4



print(f"\n九 CA_score calculado para {df_final_limpio['CA_Score'].notna().sum():,} registros")
```
